```python
N = 256
```

```python
h, w = img.shape[:2]
w2h = lambda x: int(x * h / w)
h2w = lambda x: int(x * w / h)
```

```python
center_crop = CenterCrop(height=w2h(N), width=N)
```

```python
center_pipeline = Compose([center_crop, final_padding], p=1.)
```

```python
rotate = albumentations.Rotate(limit=30, p=1.)
```

```python
random_crop = albumentations.RandomCrop(w2h(N), N, p=1.)
```

```python
# scalling to N by max side
down_scalling = albumentations.LongestMaxSize(N)
# big rotation
rotation = Rotate(limit=30, p=1.)
# pads image to be N x N
padding = PadIfNeeded(min_height=N, min_width=N, border_mode=cv2.BORDER_CONSTANT)
# small angle rotation
small_rotation = Rotate(15, p=1.)
# big rotation pipeline
rotate_pipeline = Compose([rotation, down_scalling, padding], p=0.28571429)
# no augmentations pipeline
default_pipeline = Compose([down_scalling, padding], p=0.5)

def augment_image(image, mask=None, N=256):
    #height_and_width = image.shape[:2] XXX - figure out what's with vertical images
    #h, w = min(height_and_width), max(height_and_width)
    h, w = image.shape[:2]
    
    new_height = int(round(N * h / w))
    # crops center of image
    center_crop = CenterCrop(height=new_height, width=N)
    # crops patch of random size
    random_size_crop = RandomSizedCrop((new_height, h), new_height, N, w2h_ratio=w/h)
    # crops path of constant size
    random_crop = albumentations.RandomCrop(new_height, N, p=1.)
    
    center_pipeline = Compose([center_crop, padding], p=0.14285714)
    random_size_crop_pipeline = Compose([random_size_crop, padding], p=0.28571429)
    rotated_crop_pipeline = Compose([small_rotation, center_crop, padding], p=0.14285714)
    random_crop_pipeline = Compose([random_crop, padding], p=0.14285714)
    
    augmentation = OneOf([rotate_pipeline, center_pipeline, random_size_crop_pipeline, rotated_crop_pipeline, random_crop_pipeline, default_pipeline], p=1.)
    return augmentation(image=image)['image']
```

```python
for i in range(50):
    cv2.imwrite(os.path.join('../sandbox/12-12-augmentations-test', f'{i}.jpg'), augment_image(img))
```
