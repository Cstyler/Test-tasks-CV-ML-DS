```python
class InstanceSegmentationDataset(Dataset):
    def __init__(self, dataset_dir, transform):
        image_files_path = list(filter(lambda x: x.endswith('.jpg'),
                                       os.listdir(os.path.join(dataset_dir, 'images'))))
        self.image_path_list = [os.path.join(dataset_dir, 'images', path) for path in image_files_path]
        self.mask_path_list = [os.path.join(dataset_dir, 'labels', '{0}.npy'.format(path.split('.')[0])) for path in image_files_path]
        self.transform = transform
        
    def __len__(self):
        return len(self.image_path_list)
    
    def __getitem__(self, idx):
        image_path = self.image_path_list[idx]
        with open(image_path, 'rb') as f:
            img = jpeg.decode(f.read())
        mask_path = self.mask_path_list[idx]
        with open(mask_path, 'rb') as f:
            mask = np.load(f)
        
        if self.transform:
            return self.transform(img, mask)
        else:
            return {'image': img, 'mask': mask}
```

```python
class MarketDataset(Dataset):
    def __init__(self, market, dataset_dir, market2ids):
        existing_stems = [stem
                          for stem in market2ids[market]
                          if os.path.isfile(os.path.join(dataset_dir, 'images', stem + '.jpg')) and \
                             os.path.isfile(os.path.join(dataset_dir, 'labels', stem + '.npy'))]
        
        self.image_path_list = [os.path.join(dataset_dir, 'images', stem + '.jpg') for stem in existing_stems]
        self.mask_path_list = [os.path.join(dataset_dir, 'labels', stem + '.npy') for stem in existing_stems]
        self.transform = lambda image, mask: keep_pipeline(image=image, mask=mask)
        
    def __len__(self):
        return len(self.image_path_list)
    
    def __getitem__(self, idx):
        image_path = self.image_path_list[idx]
        try:
            with open(image_path, 'rb') as f:
                img = jpeg.decode(f.read())
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            mask_path = self.mask_path_list[idx]
            with open(mask_path, 'rb') as f:
                mask = np.load(f)
        except FileNotFoundError as e:
            print('not found', e)
        
        if self.transform:
            result = self.transform(img, mask)
            return result
        else:
            return {'image': img, 'mask': mask}
```

```python
def get_background(mask):
    # np array h x w x 1
    # returns last channel (background) for the mask
    return (mask == 0).astype(np.uint8) * 255

def convert_0class_to_9classes(mask):
    # np array h x w
    mask = np.expand_dims(mask, axis=-1)
    mask8 = np.unpackbits(mask, axis=-1) * 255
    return np.concatenate((mask8, get_background(mask)), axis=2)
```

```python
N = 256
padding = PadIfNeeded(min_height=N, min_width=N, border_mode=0, always_apply=True, p=1.)
normalize = Normalize()
to_tensor = AT.ToTensor(num_classes=9)
norm_and_torch = Compose([normalize, to_tensor], p=1.)

def preprocess(image, mask):
    dict_ = padding(image=image, mask=mask)
    mask = convert_0class_to_9classes(dict_['mask'])
    return norm_and_torch(image=dict_['image'], mask=mask)
```

```python
# normalize with ImageNet values
normalize = Normalize()
# scalling to N by max side
down_scalling = LongestMaxSize(N, always_apply=True, p=1.)
# big rotation
rotation = Rotate(limit=30, border_mode=3, p=1.)
# pads image to be N x N
padding = PadIfNeeded(min_height=N, min_width=N, border_mode=0, always_apply=True, p=1.)
# small angle rotation
small_rotation = Rotate(15, p=1.)
# convert in pyTorch format
to_tensor = AT.ToTensor(num_classes=5)
# big rotation pipeline
rotate_pipeline = Compose([rotation, down_scalling, padding], p=0.28571429)
# no augmentations pipeline
default_pipeline = Compose([down_scalling, padding], p=0.5)
# for keep
keep_pipeline = Compose([default_pipeline, normalize, to_tensor], p=1.0)

def augment(image, mask=None, N=256):
    h, w = image.shape[:2]
    
    if h > w:
        new_width = int(round(N * w / h))
        # crops center of image
        center_crop = CenterCrop(height=N, width=new_width)
        # crops patch of random size
        random_size_crop = RandomSizedCrop((N, h), N, new_width, w2h_ratio=w/h)
        # crops path of constant size
        random_crop = RandomCrop(N, new_width, p=1.)
    else:
        new_height = int(round(N * h / w))
        # crops center of image
        center_crop = CenterCrop(height=new_height, width=N)
        # crops patch of random size
        random_size_crop = RandomSizedCrop((new_height, h), new_height, N, w2h_ratio=w/h)
        # crops path of constant size
        random_crop = RandomCrop(new_height, N, p=1.)
    
    center_pipeline = Compose([center_crop, padding], p=0.14285714)
    random_size_crop_pipeline = Compose([random_size_crop, padding], p=0.28571429)
    rotated_crop_pipeline = Compose([small_rotation, center_crop, padding], p=0.14285714)
    random_crop_pipeline = Compose([random_crop, padding], p=0.14285714)
    
    augmentation = Compose([OneOf([rotate_pipeline, center_pipeline, random_size_crop_pipeline, rotated_crop_pipeline, random_crop_pipeline, default_pipeline], p=1.),
                            normalize, to_tensor])
    if mask is None:
        return augmentation(image=image)['image']
    else:
        return augmentation(image=image, mask=mask)
    
def keep(image, mask=None, N=256):
    if mask is None:
        return keep_pipeline(image=image)['image']
    else:
        return keep_pipeline(image=image, mask=mask)
```

```python
class ProductLoss:
    eps = 1e-8
    product_code_channel = 1
    background_channel = 8
    weight_product_code = 2.
    def __init__(self, nll_weight=1, dice_weight=2):
        self.nll_loss = nn.BCEWithLogitsLoss(reduction='sum')
        self.weight_nll = nll_weight
        self.weight_dice = dice_weight
        
        self.non_product_code_torch_indecies = self.__make_torch_indecies(filter(lambda x: x != ProductLoss.product_code_channel,
                                                                                 range(9)))
        self.product_code_torch_indecies = self.__make_torch_indecies([ProductLoss.product_code_channel])
        self.non_product_code_and_backgorund_indecies = self.__make_torch_indecies(filter(lambda x: x != self.product_code_channel \
                                                                                                and x != self.background_channel,
                                                                                   range(9)))
    
    def __call__(self, outputs, targets):
        total_elements_number = targets.size().numel()
        (targets_no_product_code, outputs_no_product_code,
                targets_with_product_code, outputs_with_product_code) = self._product_code_split_by_emptiness(targets, outputs)
        loss = cuda(torch.tensor(0., requires_grad=True))
        #print('initial loss' + str(loss))
        dice_loss = cuda(torch.tensor(0., requires_grad=True))
        denominator = 0
        if len(targets_with_product_code):
            targets_with_product_code_product_code = self._select_channels(targets_with_product_code, 
                                                                           self.product_code_torch_indecies)
            outputs_with_product_code_product_code = self._select_channels(outputs_with_product_code,
                                                                           self.product_code_torch_indecies)
            
            targets_with_product_code_non_product_code = self._select_channels(targets_with_product_code,
                                                                               self.non_product_code_torch_indecies)
            outputs_with_product_code_non_product_code = self._select_channels(outputs_with_product_code,
                                                                               self.non_product_code_torch_indecies)
            
            loss = loss + (self.nll_loss(outputs_with_product_code_product_code, targets_with_product_code_product_code)
                           / total_elements_number
                           * self.weight_product_code
                           * self.weight_nll)
            #print('after existing product codes loss' + str(loss))
            if self.weight_dice:
                temp_dice1 = self.dice(self._postprocess(outputs_with_product_code_product_code), targets_with_product_code_product_code)
                #print('Dice with product productcode only', temp_dice1)
                dice_loss = dice_loss - torch.log(temp_dice1) * self.weight_product_code * self.weight_dice
                denominator += 1

            loss = loss + (self.nll_loss(outputs_with_product_code_non_product_code, targets_with_product_code_non_product_code)
                           / total_elements_number
                           * self.weight_nll)
            #print('after everythin but product code loss' + str(loss))
            if self.weight_dice:
                temp_dice2 = self.dice(self._postprocess(outputs_with_product_code_non_product_code), targets_with_product_code_non_product_code)
                #print('Dice with product non productcode: ', temp_dice2)
                dice_loss = dice_loss - torch.log(temp_dice2) * self.weight_dice
                denominator += 1

        if len(targets_no_product_code):
            targets_no_product_code_peeled = self._select_channels(targets_no_product_code,
                                                                   self.non_product_code_and_backgorund_indecies)
            outputs_no_product_code_peeled = self._select_channels(outputs_no_product_code,
                                                                   self.non_product_code_and_backgorund_indecies)
            loss = loss + (self.nll_loss(outputs_no_product_code_peeled, targets_no_product_code_peeled)
                           / total_elements_number
                           * self.weight_nll)
            #print('no backgorund and product code loss' + str(loss))
            if self.weight_dice:
                temp_dice3 = self.dice(self._postprocess(outputs_no_product_code_peeled), targets_no_product_code_peeled)
                #print('Dice no product no back: ', temp_dice3)
                dice_loss = dice_loss - torch.log(temp_dice3) * self.weight_dice
                denominator += 1
        #print()
        if dice_loss == 0.:
            return loss
        else:
            return loss + dice_loss / denominator
    
    @staticmethod
    def __make_torch_indecies(channels_nums_iterable):
        return cuda(torch.LongTensor(list(channels_nums_iterable)))
    
    @staticmethod
    def _select_channels(tensor, indecies):
        return torch.index_select(tensor, 1, indecies)
    
    @staticmethod
    def _postprocess(outputs):
        return torch.sigmoid(outputs)
    
    @classmethod
    def dice(cls, outputs, targets):
        # just classic dice realisation
        intersection = (outputs * targets).sum()
        union = outputs.sum() + targets.sum() + cls.eps
        return 2 * intersection / union
    
    def _product_code_split_by_emptiness(self, targets, outputs):
        # returns all samples, which have no product codes
        with torch.no_grad():
            product_codes_tensor = self._select_channels(targets, self.product_code_torch_indecies)
            product_codes_tensor = product_codes_tensor.sum(dim=(1,2,3))
            non_empty_product_codes_indecies = (product_codes_tensor != 0)
            empty_product_codes_indecies = (product_codes_tensor == 0)
        
        targets_no_product_code = targets[empty_product_codes_indecies]
        outputs_no_product_code = outputs[empty_product_codes_indecies]
        
        targets_with_product_code = targets[non_empty_product_codes_indecies]
        outputs_with_product_code = outputs[non_empty_product_codes_indecies]
        return (targets_no_product_code, outputs_no_product_code,
                targets_with_product_code, outputs_with_product_code)
```

```python
class Loss:
    def __init__(self, nll_weight=1, dice_weight=1):
        self.nll_loss = nn.BCEWithLogitsLoss()
        self.dice_weight = dice_weight
        self.nll_weight = nll_weight

    def __call__(self, outputs, targets):
        outputs = pl._select_channels(outputs, pl.non_product_code_torch_indecies)
        targets = pl._select_channels(targets, pl.non_product_code_torch_indecies)
        
        loss = self.nll_weight * self.nll_loss(outputs, targets)
        
        if self.dice_weight:
            outputs = torch.sigmoid(outputs)
            eps = 1e-15
            dice_target = targets#(targets == 1).float()
            dice_output = outputs
            intersection = (dice_output * dice_target).sum()
            union = dice_output.sum() + dice_target.sum() + eps

            loss -= self.dice_weight * torch.log(2 * intersection / union)

        return loss
```

```python
def threshold_batch(batch):
    n, c, h, w = batch.shape
    binary = batch > THRESHOLD
    binary[:, 1, ...] = torch.zeros((h, w), dtype=binary.dtype, device=cuda1080)
    sums = binary.sum(dim=(2,3)) < 10  # n, c
    binary[sums] = torch.zeros((h, w), dtype=binary.dtype, device=cuda1080)
    return binary.float()
```

```python
THRESHOLD = 0.3
def IoU(outputs, targets):
    outputs = torch.sigmoid(outputs)
    eps = 1e-15
    dice_target = (targets[:, :-1, ...] == 1.).float()
    dice_output = threshold_batch(outputs[:, :-1, ...])
    intersection = (dice_output * dice_target).sum()
    union = dice_output.sum() + dice_target.sum() + eps
    if union < 1.:
        print('in strange if')
        return 1.
    return 2 * intersection / union
```

```python
def find_ious(data_loader, model):
    model.eval()
    with torch.no_grad():
        ious = []
        for i, batch in enumerate(data_loader):
            inputs_test, targets_test = cuda(batch['image']), cuda(batch['mask'])
            output_tensor = model(inputs_test)
            ious.append(IoU(output_tensor, targets_test).item())
    return ious
```

```python
def product_code_iou(output_tensor, targets_tensor):
    eps = 1e-15
    dice_target = (targets_tensor[1] == 1.).float()
    dice_output = (torch.sigmoid(output_tensor[1]) > 0.3).float()
    intersection = (dice_output * dice_target).sum()
    union = dice_output.sum() + dice_target.sum() + eps
    if (union < 1.):
        return 1.
    result = 2 * intersection / union
    return result.item()
```

```python
def product_code_find_iou(data_loader, model):
    model.eval()
    with torch.no_grad():
        ious = []
        for i, batch in enumerate(data_loader):
            inputs_test, targets_test = cuda(batch['image']), cuda(batch['mask'])
            output_tensor = model(inputs_test)[0]
            ious.append(product_code_iou(output_tensor, targets_test[0]))
    return np.mean(ious)
```

```python
def get_table(ious_by_market):
    table_data = []
    for market, ious in ious_by_market.items():
        data = np.array(ious)
        min_, max_, mean, std = np.min(data), np.max(data), np.mean(data), np.std(data)
        table_data.append((market, min_, max_, mean, std))
    writer = MarkdownTableWriter()
    writer.table_name = "IoU by markets"
    writer.headers = ("market", "min", "max", "mean", "std")
    writer.value_matrix = table_data
    return writer.dumps()
```

```python
import mask2bbox
import typex
import type_utils
import typed
from operator import itemgetter

def predict(img, model, use_cuda=False, truncate_last_channel=True):
    batch_mode = (type(img) == list)
    if batch_mode:
        images_shapes = [x.shape for x in img]
        img = [process_image(x) for x in img]
        img = torch.stack(img, dim=0)
    else:
        image_shape = img.shape
        img = process_image(img)
        img = img[None,...]
    with torch.no_grad():
        if use_cuda:
            output_tensor = torch.sigmoid(model(cuda(img))).cpu().numpy()
        else:
            output_tensor = torch.sigmoid(model(img)).numpy()
    if truncate_last_channel:
        output_tensor = output_tensor[:, :-1, ...]
        
    if batch_mode:
        output_tensors = [transp(x) for x in output_tensor]
        return [crop_initial_aspect(output_tensor, image_shape)
                for output_tensor, image_shape in zip(output_tensors, images_shapes)]
    else:
        output_tensor = transp(output_tensor[0])
        return crop_initial_aspect(output_tensor, image_shape)

def crop_initial_aspect(mask, source_img_shape):
    n = mask.shape[1]
    H, W = source_img_shape[:2]
    if H < W:
        h2w = H / W
        h = n * h2w
        w = n
    else:
        w2h = W / H
        w = n * w2h
        h = n
    return mask[round(n/2 - h/2): round(n/2 + h/2), round(n/2 - w/2): round(n/2 + w/2), :]

N = 256 # images max size

# normalize with ImageNet values
normalize = Normalize()
# scalling to N by max side
down_scalling = LongestMaxSize(N, always_apply=True, p=1.)
# pads image to be N x N
padding = PadIfNeeded(min_height=N, min_width=N, border_mode=0, always_apply=True, p=1.)
# convert in pyTorch format
to_tensor = AT.ToTensor(num_classes=9)
# no augmentations pipeline
default_pipeline = Compose([down_scalling, padding], p=1.0)
# for keep
keep_pipeline = Compose([default_pipeline, normalize, to_tensor], p=1.0)
    
def process_image(image, N=256):
    return keep_pipeline(image=image)['image']

def transp(tensor):
    return np.transpose(tensor, axes=[1,2,0])

def predict_function(image):
    return predict(image, model, use_cuda=True)

def imitate_tag(tag_id, segs_by_type_machine, segs_by_type_manual):
    new_machine = type_utils.segmarker2elastic(segs_by_type_machine)
    new_human = type_utils.segmarker2elastic_human(segs_by_type_manual)
    if not segs_by_type_machine:
        new_machine = []
    return {
        '_id': tag_id,
        '_source': {
            'segments': new_machine + new_human
        }
    }

def agg_segments_by_type(segments):
    return {x['segmentType']: x['coordinates'] for x in segments}

def calculate_accuracy_by_markets(predict_function, processor, base_dir, important_markets_set):
    # predict_function(full_image) -> layout
    markets_dirs = [(x, os.path.join(base_dir, x)) for x in os.listdir(base_dir) if not x.startswith('.')]
    market_scores = {}
    for market, market_directory in markets_dirs:
#         if market not in important_markets_set:
#             continue
        if market == 'billa.ru' or market == 'karusel.ru':
            continue
        layouts_dir = os.path.join(market_directory, 'layout')
        images_dir = os.path.join(market_directory, 'images')
        tag_ids = {x.split('.')[0] for x in os.listdir(layouts_dir) if not x.startswith('.') and x.endswith('.json')} & \
                  {x.split('.')[0] for x in os.listdir(images_dir) if not x.startswith('.') and x.endswith('.jpg')}

        tag_ids_list, images, images_shapes, human_layouts_list = [], [], [], []
        for tag_id in tag_ids:
            tag_ids_list.append(tag_id)
            layout_path = os.path.join(layouts_dir, tag_id) + '.json'
            image_path = os.path.join(images_dir, tag_id) + '.jpg'
            image = cv2.imread(image_path)
            image, image_shape = cv2.cvtColor(image, cv2.COLOR_BGR2RGB), image.shape
            images.append(image)
            images_shapes.append(image_shape)
            with open(layout_path, 'r') as f:
                human_layout = json.load(f)
                _, human_layout = typed.get_segments({"_source": {"segments": human_layout}})
            human_layouts_list.append(agg_segments_by_type(human_layout))

        masks_list = []
        for i in range(0, len(images)-1, 2):
            masks_list += predict_function(images[i: i+2])
        #masks_list = predict_function(images)
        machine_layouts_list = processor.mask2rectangles(masks_list, images_shapes, threshold=0.01)
        tag_dict = {}
        type_d_tags = []
        for tag_id, machine_segments, manual_segments in zip(tag_ids_list, machine_layouts_list, human_layouts_list):
            tag_dict[tag_id] = {}
            tag_dict[tag_id]['machineSegments'] = machine_segments
            tag_dict[tag_id]['manualSegments'] = manual_segments
            type_d_tags.append(imitate_tag(tag_id, machine_segments, manual_segments))

        market_scores[market] = (type_utils.mean_of_iterable(map(itemgetter(1), typex.calc_metrics(tag_dict))),
                                 typed.type_d_mean_over_types_and_tags(type_d_tags).recall)
    return market_scores

processor = mask2bbox.Processor()
```

```python
import itertools
from random import shuffle

palette = list(itertools.product(np.linspace(0, 255, 10).astype(np.uint8),
                                 repeat=3))
shuffle(palette)
palette = [np.array(p, dtype=int) for p in palette]
```

```python
def transp(tensor):
    return np.transpose(tensor, axes=[1,2,0])
```

```python
IMAGENET_MEAN = (0.485, 0.456, 0.406)
IMAGENET_STD = (0.229, 0.224, 0.225)

def revert_image(image):
    return cv2.cvtColor(((transp(image) * IMAGENET_STD + IMAGENET_MEAN)*255).astype(np.uint8), cv2.COLOR_BGR2RGB)
```

```python
def output_tensor2pic(output_tensor):
    # 9, 256, 256
    mask = transp(output_tensor)
    # 256, 256, 9
    channels_num = mask.shape[-1] - 1  # we're not interested in background class
    result = np.ndarray((channels_num, mask.shape[0], mask.shape[1], 3), dtype=np.uint8)  #8, 256, 256, 3
    for channel in range(channels_num):
        result[channel, ...] = (np.expand_dims(mask[..., channel], -1) * palette[channel].reshape(1,1,3)).astype(np.uint8)
    return np.sum(result, axis=0, dtype=np.uint8)
```

```python
def batch2pics(output_tensor, label_tensor, image_tensor):
    processed_batch = []
    batch_size = output_tensor.shape[0]
    for i in range(batch_size):
        mask = output_tensor2pic(output_tensor[i])
        gt = output_tensor2pic(label_tensor[i])
        image = revert_image(image_tensor[i])
        processed_batch.append(dict(
            mask = mask,
            gt = gt,
            image = image
        ))
    return processed_batch
```

```python
def get_plot(batch, id_=0, prefix='', dir_='./results'):
    n = len(batch)
    plt.ioff()
    f, axarr = plt.subplots(n, 5, figsize=(30,50))

    if n > 1:
        for i in range(n):
            image = batch[i]['image']
            label = batch[i]['gt']
            mask = batch[i]['mask']
            axarr[i, 0].imshow(image)
            axarr[i, 1].imshow(label)
            axarr[i, 2].imshow(mask)
            axarr[i, 3].imshow(blend(image, label))
            axarr[i, 4].imshow(blend(image, mask))
    else:
        image = batch[i]['image']
        label = batch[i]['gt']
        mask = batch[i]['mask']
        axarr[0].imshow(image)
        axarr[1].imshow(label)        
        axarr[2].imshow(mask)
        axarr[3].imshow(blend(image, label))
        axarr[4].imshow(blend(image, mask))
    plt.savefig(f'{dir_}/{prefix}grid{id_}.jpg')
    plt.close(f)
```

```python
def save_results(data_loader, model, batches_num, prefix='', on_gpu=True, postprocess_batch_func=lambda x: x):
    processed_batches = []
    model.eval()
    print('Inference...')
    with torch.no_grad():
        for i, test_batch in tqdm_notebook(enumerate(data_loader)):
            if i >= batches_num:
                break
            inputs_test, targets_test = test_batch['image'], test_batch['mask']
            if on_gpu:
                output_tensor = torch.sigmoid(model(cuda(inputs_test))).cpu().numpy()
            else:
                output_tensor = torch.sigmoid(model(inputs_test)).numpy()
            label_tensor = targets_test.numpy()
            image_tensor = inputs_test.numpy()
            output_tensor = postprocess_batch_func(output_tensor)
            processed_batches.append(batch2pics(output_tensor, label_tensor, image_tensor))
    print('Saving...')
    for i, proc_batch in tqdm_notebook(enumerate(processed_batches)):
        get_plot(proc_batch, i, prefix)
    print('Done!')
```

```python
def threshold_batch(batch):
    n, c, h, w = batch.shape
    binary = batch > 0.1
    binary[:, 1, ...] = np.zeros((h, w), dtype=binary.dtype)
    sums = binary.sum(axis=(2,3)) < 10  # n, c
    binary[sums] = np.zeros((h, w), dtype=binary.dtype)
    return binary.astype(np.float32)
```

## Assess and Compare speed of CPU vs GPU

```python
def get_dataloader_with_batchsize_n(n):
    return torch.utils.data.DataLoader(InstanceSegmentationDataset(val_dir, keep), batch_size=n, shuffle=True, num_workers=2)
```

```python
def k_iterations(model, k, batch_size, use_gpu):
    if use_gpu:
        model = model.cuda()
    else:
        model = model.cpu()
    model.eval()
    with torch.no_grad():
        for _ in tqdm_notebook(range(k), total=k):
            dataloader = get_dataloader_with_batchsize_n(batch_size)
            for batch in dataloader:
                inputs_test, targets_test = batch['image'], batch['mask']
                if use_gpu:
                    output_tensor = torch.sigmoid(model(cuda(inputs_test))).cpu().numpy()
                else:
                    output_tensor = torch.sigmoid(model(inputs_test)).numpy()
```

```python
DATASET_LEN = len(InstanceSegmentationDataset(val_dir, keep))
print(DATASET_LEN)
```

```python
K = 1
```

```python
print('GPU batch 8 average per image:')
tic = time.time()
k_iterations(model, K, 8, True)
execution_duration = time.time() - tic
print(execution_duration / DATASET_LEN / K)

print('GPU batch 16 average per image:')
tic = time.time()
k_iterations(model, K, 16, True)
execution_duration = time.time() - tic
print(execution_duration / DATASET_LEN / K)

print('CPU batch 8 average per image:')
tic = time.time()
k_iterations(model, K, 8, False)
execution_duration = time.time() - tic
print(execution_duration / DATASET_LEN / K)

print('CPU batch 16 average per image:')
tic = time.time()
k_iterations(model, K, 16, False)
execution_duration = time.time() - tic
print(execution_duration / DATASET_LEN / K)
```
